```python
# ======= Mount Google Drive =======
from google.colab import drive
drive.mount('/content/drive')

# ======= Configuration =======
import os
BASE_DIR = '/content/drive/MyDrive/DATAP'   # <--- your folder
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
SEED = 42
EPOCHS = 25      # change as needed
MODEL_SAVE_PATH = '/content/drive/MyDrive/pisoriasis_mobilenetv2_best.h5'

# quick check of folder structure
print("Contents of dataset folder:")
for d in os.listdir(BASE_DIR):
    print(" -", d)
print()

# ======= Imports =======
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils import class_weight

# ======= Data generators with validation split =======
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.10,
    height_shift_range=0.10,
    shear_range=0.05,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2  # 20% for validation
)

train_generator = train_datagen.flow_from_directory(
    BASE_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=True,
    subset='training',
    seed=SEED
)

valid_generator = train_datagen.flow_from_directory(
    BASE_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False,
    subset='validation',
    seed=SEED
)

# ======= Optional: compute class weights to handle imbalance =======
# train_generator.classes gives numeric labels
y_train = train_generator.classes
class_weights = class_weight.compute_class_weight(class_weight='balanced',
                                                  classes=np.unique(y_train),
                                                  y=y_train)
class_weights = {i: w for i, w in enumerate(class_weights)}
print("Class weights:", class_weights)

# ======= Build model: MobileNetV2 backbone + custom head =======
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
base_model.trainable = False   # freeze base initially

# Head
inputs = tf.keras.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(1, activation='sigmoid')(x)  # binary
model = models.Model(inputs, outputs)

model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ======= Callbacks =======
callbacks = [
    EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True, verbose=1),
    ModelCheckpoint(MODEL_SAVE_PATH, monitor='val_loss', save_best_only=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1, min_lr=1e-7)
]

# ======= Training - Stage 1 (train head) =======
steps_per_epoch = max(1, train_generator.samples // BATCH_SIZE)
validation_steps = max(1, valid_generator.samples // BATCH_SIZE)

history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    validation_data=valid_generator,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    class_weight=class_weights,
    callbacks=callbacks
)

# ======= Optional Stage 2: fine-tune some of the base_model =======
# Unfreeze last N layers of base model and continue training with lower LR
UNFREEZE_AT = 100  # layer index to start unfreezing; tune as needed

base_model.trainable = True
# Freeze all layers before UNFREEZE_AT
for layer in base_model.layers[:UNFREEZE_AT]:
    layer.trainable = False
for layer in base_model.layers[UNFREEZE_AT:]:
    layer.trainable = True

# recompile with lower LR
model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-5),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# continue training (fine-tune)
fine_tune_epochs = 10
total_epochs = EPOCHS + fine_tune_epochs

history_fine = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    validation_data=valid_generator,
    validation_steps=validation_steps,
    epochs=total_epochs,
    initial_epoch=history.epoch[-1] if hasattr(history, 'epoch') else 0,
    class_weight=class_weights,
    callbacks=callbacks
)

# ======= Save final model (overwrite best file) =======
model.save('/content/drive/MyDrive/pisoriasis_mobilenetv2_final.h5')
print("Saved final model to /content/drive/MyDrive/pisoriasis_mobilenetv2_final.h5")

# ======= Plot training curves =======
def plot_history(h1, h2=None):
    h = {}
    # merge histories
    for k in h1.history:
        h[k] = h1.history[k]
    if h2:
        for k in h2.history:
            if k in h:
                h[k] = h[k] + h2.history[k]
            else:
                h[k] = h2.history[k]
    epochs = range(1, len(h['loss'])+1)
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(epochs, h['loss'], label='train_loss')
    plt.plot(epochs, h['val_loss'], label='val_loss')
    plt.title('Loss')
    plt.legend()
    plt.subplot(1,2,2)
    plt.plot(epochs, h['accuracy'], label='train_acc')
    plt.plot(epochs, h['val_accuracy'], label='val_acc')
    plt.title('Accuracy')
    plt.legend()
    plt.show()

plot_history(history, history_fine if 'history_fine' in globals() else None)

# ======= Evaluate on validation set and show confusion matrix & classification report =======
# Build predictions on validation set
valid_generator.reset()
preds = model.predict(valid_generator, steps=validation_steps, verbose=1)
y_pred = (preds.ravel() >= 0.5).astype(int)

# Because validation generator may not include all images if steps < samples/BATCH, gather true labels accordingly:
y_true = valid_generator.classes[:len(y_pred)]

print("\nClassification report:")
print(classification_report(y_true, y_pred, target_names=list(valid_generator.class_indices.keys())))

cm = confusion_matrix(y_true, y_pred)
print("Confusion matrix:\n", cm)

# ======= Quick inference helper function =======
from tensorflow.keras.preprocessing import image
def predict_image(img_path, model, target_size=IMG_SIZE):
    img = image.load_img(img_path, target_size=target_size)
    x = image.img_to_array(img) / 255.0
    x = np.expand_dims(x, axis=0)
    p = model.predict(x)[0][0]
    return p  # probability of class '1' (depends on mapping)

# Show mapping of class indices
print("Class indices:", valid_generator.class_indices)
```

```python
from tensorflow.keras.models import load_model, Model, Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
import numpy as np
import math

# ====== 1. Load trained MobileNetV2 model ======
mobilenet = load_model("/content/drive/MyDrive/pisoriasis_mobilenetv2_final.h5")

# ====== 2. Create feature extractor ======
# Use the layer just before the final classification layer
# Usually it's 'global_average_pooling2d' or check mobilenet.summary() to confirm
feature_layer_name = "global_average_pooling2d"  # replace with actual layer name
feature_extractor = Model(inputs=mobilenet.input,
                          outputs=mobilenet.get_layer(feature_layer_name).output)
feature_extractor.summary()

# ====== 3. Extract features from validation generator ======
# Ensure valid_generator is already defined and uses shuffle=False
valid_generator.shuffle = False
steps = math.ceil(valid_generator.samples / valid_generator.batch_size)
features = feature_extractor.predict(valid_generator, steps=steps, verbose=1)
labels = valid_generator.classes[:features.shape[0]]  # align labels

print("Feature shape:", features.shape)  # e.g., (num_samples, 1280) if GAP used

# ====== 4. Reshape for LSTM ======
# If features are 2D (num_samples, 1280), reshape to (num_samples, timesteps, features_per_step)
# We can treat each feature vector as a single timestep
X = features.reshape((features.shape[0], 1, features.shape[1]))
print("Reshaped for LSTM:", X.shape)  # (num_samples, 1, 1280)

# ====== 5. Define LSTM model ======
lstm = Sequential([
    LSTM(64, input_shape=(X.shape[1], X.shape[2])),
    Dense(1, activation='sigmoid')
])
lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
lstm.summary()

# ====== 6. Train LSTM ======
lstm.fit(X, labels, epochs=5, batch_size=32, validation_split=0.2, verbose=1)

# ====== 7. Save LSTM model ======
lstm.save("/content/drive/MyDrive/lstm_model_final.h5")
print("✅ LSTM model saved successfully!")
```
